```python
#import 

from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
#Load data
netflix_stocks = pd.read_csv("NFLX.csv")
netflix_stocks.head()
```

```python
#Inspect
dowjones_stocks = pd.read_csv("DJI.csv")
dowjones_stocks.head()
```

```python
#Rename column
netflix_stocks.rename(columns={"Adj Close": "Price"}, inplace=True)
dowjones_stocks.rename(columns={"Adj Close": "Price"}, inplace=True)
netflix_stocks.head()
```

```python
#Netflix stock price distribution 2017
ax = sns.violinplot(netflix_stocks['Price'], color="red")
ax.set_title("Distribution of 2017 Netflix Stock Price")
plt.show()
```

```python
#Earnings per share in Cents
x_positions = [1, 2, 3, 4]
chart_labels = ["1Q2017","2Q2017","3Q2017","4Q2017"]
earnings_actual =[.4, .15,.29,.41]
earnings_estimate = [.37,.15,.32,.41 ]
plt.scatter(x_positions, earnings_actual,  color="red",alpha=0.5)
plt.scatter(x_positions, earnings_estimate, color="blue",alpha=0.5)
plt.xticks(x_positions, chart_labels)
plt.legend(["Actual", "Estimate"])
plt.title("Earnings Per Share in Cents")
plt.show()
```

```python
#Revenue and earnings
# The metrics below are in billions of dollars
revenue_by_quarter = [2.79, 2.98,3.29,3.7]
earnings_by_quarter = [.0656,.12959,.18552,.29012]
quarter_labels = ["2Q2017","3Q2017","4Q2017", "1Q2018"]

# Revenue
n = 1  # This is our first dataset (out of 2)
t = 2 # Number of dataset
d = 4 # Number of sets of bars
w = 0.8 # Width of each bar
bars1_x = [t*element + w*n for element
             in range(d)]
plt.bar(bars1_x, revenue_by_quarter)


# Earnings
n = 2  # This is our second dataset (out of 2)
t = 2 # Number of dataset
d = 4 # Number of sets of bars
w = 0.8 # Width of each bar
bars2_x = [t*element + w*n for element
             in range(d)]
plt.bar(bars2_x, earnings_by_quarter)

middle_x = [ (a + b) / 2.0 for a, b in zip(bars1_x, bars2_x)]
plt.xticks(middle_x, quarter_labels)
print(middle_x)
labels = ["Revenue", "Earnings"]
plt.legend(labels)
plt.title("Netflix Revenue and Earnings in $ Billions")
plt.show()
```

```python
#Compare Netflix Dow Jones
fig = plt.figure(figsize=(8, 3), dpi=100)
plt.subplots_adjust(hspace=.5, wspace=0.5)
# Left plot Netflix
ax1 = plt.subplot(1, 2, 1)
ax1.set_xlabel('Date')
ax1.set_ylabel('Stock Price')
ax1.set_title("Netflix")
plt.xticks(rotation='vertical')
plt.plot(netflix_stocks['Date'], netflix_stocks['Price'], color="purple")

# Right plot Dow Jones
ax2 = plt.subplot(1, 2, 2)
ax2.set_xlabel('Date')
ax2.set_ylabel('Stock Price')
ax2.set_title("Dow Jones")
plt.xticks(rotation='vertical')
plt.plot(dowjones_stocks['Date'], dowjones_stocks['Price'], color="green")
plt.show()
```

```python
#Overview plot
fig = plt.figure(figsize=(11, 10), dpi=100)
plt.subplots_adjust(hspace=.45, wspace=0.25)
ax0 = fig.add_subplot(311)
ax1 = fig.add_subplot(323)
ax2 = fig.add_subplot(324)
ax3 = fig.add_subplot(325)
ax4 = fig.add_subplot(326)

#0
sns.violinplot(netflix_stocks['Price'], color="blue", ax=ax0, alpha=1)
ax0.set_title("Distribution of 2017 Netflix Stock Price")
plt.setp(ax0.collections, alpha=.7)

#1
ax1.scatter(x_positions, earnings_actual,  color="red",alpha=0.5, s=100)
ax1.scatter(x_positions, earnings_estimate, color="blue",alpha=0.5, s=100)
ax1.set_xticks(x_positions, chart_labels)
ax1.legend(["Actual", "Estimate"])
ax1.set_title("Earnings Per Share in Cents")

#2
ax2.bar(bars1_x, revenue_by_quarter)
ax2.bar(bars2_x, earnings_by_quarter)
ax2.set_xticks(middle_x)
ax2.set_xticklabels(quarter_labels)
ax2.legend(labels, loc='upper left')
ax2.set_title("Netflix Revenue and Earnings in $ Billions")

#3
ax3.plot(netflix_stocks['Date'], netflix_stocks['Price'], color="purple", lw=3)
ax3.set_xlabel('Date')
ax3.set_ylabel('Stock Price')
ax3.set_title("Netflix")
ax3.tick_params(axis='x', rotation=90)

#4
ax4.plot(dowjones_stocks['Date'], dowjones_stocks['Price'], color="green", lw=3)
ax4.set_xlabel('Date')
ax4.set_ylabel('Stock Price')
ax4.set_title("Dow Jones")
ax4.tick_params(axis='x',rotation=90)

for plot in[ax0, ax1, ax2, ax3, ax4]:
    for axis in ['bottom','left','right', 'top']:
        plot.spines[axis].set_linewidth(1.5)
        plot.tick_params(direction='in')
```
